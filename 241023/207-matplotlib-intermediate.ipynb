```python
import matplotlib.pyplot as plt

plt.plot([10,20,30,40], 'r.', label='circle')
plt.plot([40,30,20,10],'g^:', label='Triangle Up')

plt.title('marker')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3,4],[10,20,30,40],'-m*',[1,2,3,4],[50,60,70,80], 'r.',label = 'circle')
plt.plot([40,30,20,10],'g^', label = 'trianle up')
plt.title('marker')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.show()
```

```python
plt.plot([10,20,30,40],[1,4,9,16],c='b',lw=5, ls='--', marker='o', mec='g', mew=5, mfc='r', ms=15)
plt.title("Style")
plt.show()
```
